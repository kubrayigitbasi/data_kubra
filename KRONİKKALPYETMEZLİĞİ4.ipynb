```python
# Kütüphanelerin Çağırılması
import numpy as np
import pandas as pd

import seaborn as sns
sns.set()

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Veriyi okumadan önce Sütun isimlerini içerisinde barındıran bir liste oluşturduk, kendi istediğimiz gibi isimlendirdik.
features = ["ID" , "Center_cat" , "Age" , "Gender" , "Weight" , "Height" , "BMI" , "NYHA" , "FE" , "VO2_peak" , "VO2_peak_Kg" , "VO2_peak_perc" , "VE_VCO2_S" , "MDRD" , "Crea" , "Na" , "K" , "Hb" , "Ac_Urico" , "FU_total" , "Exitus" , "causa" , "ACE_I" , "AT1_I" , "BB" , "Tipo BB" , "Dose BB" , "Dose_equiv" , "Diuretici" , "Diur_tipo" , "Dose diur" , "Diur2_tipo" , "Dose diur1" , "Statin" , "Allopurinolo" , "Risparmiatore_K" , "Antiaggreganti" , "TAO" , "Digitale" , "Amiodarone" , "eziologia"]


# Veriyi okurken names= features diyerek kendi veridiğimiz isimlere göre okuması için komut verdik.
KKY = pd.read_csv("https://raw.githubusercontent.com/kubrayigitbasi/data_kubra/main/datakubra.csv", names= features, header=0)
KKY.sample(5)
```

```python
# Veri setinin boyutları
print("Satır:", KKY.shape[0], '\t', "Sütun:", KKY.shape[1])

# bir Satır boşluk bırak
print("\n")
print("\t")

# Veri setinin sütun isimleri
print(KKY.columns)
```

```python
# Veriseti hakkında genel bilgi 
KKY.info()
```

```python
# replace() : verileri değiştirmek için kullanırız.
KKY = KKY.replace({'Exitus': {"no risk": 0, "yes risk": 1,}})
KKY.sample(7)
```

```python
# Veriyi istatistiksel açıdan inceleyelim.
KKY.describe()
```
